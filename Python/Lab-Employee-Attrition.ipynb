```python
# Import pandas 
import pandas as pd
import seaborn as sns
```

```python
# Load employee_attrition.csv
df=pd.read_csv('employee_attrition.csv')
```

```python
df
```

```python
# Descriptive statistics
# Attrition distribution
df['attrition'].value_counts()
```

```python
# Plot: Attrition distribution
sns.countplot(data=df, x='attrition')
```

```python
# Group by 'OverTime' and 'BusinessTravel' for attrition rates
# Group by 'OverTime' and calculate attrition rate
df.groupby('OverTime')['attrition'].value_counts(normalize=True)*100
```

```python
# Group by 'BusinessTravel' and calculate attrition rate
df.groupby('BusinessTravel')['attrition'].value_counts(normalize=True)*100
```

```python
# Countplot for Gender vs Attrition
sns.countplot(data=df, x='gender', hue='attrition')
```

```python
# Countplot for Marital Status vs Attrition
sns.countplot(data=df, x='MaritalStatus', hue='attrition')
```

```python
# Histogram for Age with hue for Attrition
sns.histplot(data=df, x='age', hue='attrition', multiple='stack')
```

```python
# Scatter plot of Age vs. Total Working Years, colored by Attrition
sns.scatterplot(data=df, x='age', y='TotalWorkingYears', hue='attrition')
```

```python
# Scatter plot of DailyRate vs. DistanceFromHome, colored by Attrition
sns.scatterplot(data=df, x='DailyRate', y='DistanceFromHome', hue='attrition')
```

```python
# Scatter plot of NumCompaniesWorked vs. Total Working Years, colored by Attrition
sns.scatterplot(data=df, x='NumCompaniesWorked', y='TotalWorkingYears', hue='attrition')
```
