- `Time`: Purchase time (10 am to 9 pm)
- `Payment`: Payment used by the customer for the purchase (3 methods are available – Cash, Credit card and Ewallet)
- `COGS`: Cost of goods sold
- `Gross margin percentage`: Gross margin percentage
- `Gross income`: Gross income
- `Rating`: Customer stratification rating on their overall shopping experience (On a scale of 1 to 10)

#### Average Total Price of Sales in Mandalay City with a Customer Rating Above 8

```python
SM_mandalayrating=SM[(SM['City'] == 'Mandalay') & (SM['Rating'] > 8)]
```

```python
SM_mandalayrating
```

```python
SM_mandalayrating['Total'].mean()
```

#### Product Line Contribution to Total Sales Amount

```python
product_line_sales = SM.groupby('Product line').aggregate({'Total':'sum'})
```

```python
product_line_sales['Total'].sort_values()
```

####  Number of Sales by Payment Method in Yangon

```python
yangon_payment_methods= SM[(SM['City'] =='Yangon')]
```

```python
yangon_payment_methods['Payment'].value_counts()
```

#### Average Gross Income by Customer Type

```python
SM.groupby('Customer type')['gross income'].mean()
```
