```python
import pandas as pd
```

```python
EM=pd.read_csv('employees_m.csv')
```

```python
EM
```

```python
EM['ave_logintime_pl']=EM['total login time']/EM['# of logins']
```

```python
EM['Bonus']=EM['salary']*EM['bonus_percent']/100
```

```python
EM['Hours']=pd.to_datetime(EM['last_login']).dt.hour
```

```python
bins=[-1,6,12,18,24]
```

```python
labels=['Night','Mornig','Afternoon','Evening']
```

```python
EM['Time Category']=pd.cut(EM['Hours'],bins=bins, labels=labels, right=False)
```

```python
EM['Time Category'].value_counts()
```
