```python
SS_agg=SS.groupby('Customer Name').aggregate({'Sales':'sum','Profit':'sum'})
```

- Q:  How does discount affect profitability?
- A: The correlation between Discount and Profit is approximately -0.22. This indicates a weak negative relationship, suggesting that as the discount increases, profitability tends to decrease slightly. While this correlation is not strong, it does imply that higher discounts are generally associated with lower profits, potentially due to reduced margins or increased losses at higher discount levels

```python
Discount_analysis = SS.groupby('Discount').aggregate({'Sales': 'sum', 'Profit': 'sum'})
```

```python
Discount_analysis
```

```python
# Calculate the correlation between Discount and Profit to analyze the relationship
correlation = SS[['Discount', 'Profit']].corr()
```
