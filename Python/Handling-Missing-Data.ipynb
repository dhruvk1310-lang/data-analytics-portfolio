Answer the following questions about `covidtotals.csv`, which contains data about covid cases and deaths. Include your final code below. When you are done, click on `Run` > `Restart Kernel and Run All Cells...` and verify the output. 
- Q: What value did you use to replace missing values in `gdp_per_capita` and how did you derive that value?
- A: To get the value i used CT['gdp_per_capita'].median() The value is 12265.791000000001
- Q: What value did you use to replace missing values in `aged_65_older` and how did you derive that value?
- A: To get the value i used CT['aged_65_older'].median() the value is 6.224  
- Q: What value did you use to replace missing values in `region` and how did you derive that value?
- A: To get the value i used CT['region'].value_counts() and the value was'Eastern Europe'
  '

```python
CT= pd.read_csv('covidtotals.csv')
```

```python
CT
```

```python
CT.isna().sum()
```

```python
CT['gdp_per_capita'].median()
```

```python
CT['gdp_per_capita'].fillna(CT['gdp_per_capita'].median())
```

```python
CT['aged_65_older'].median()
```

```python
CT['aged_65_older'].fillna(CT['aged_65_older'].median())
```

```python
CT['region'].value_counts()
```

```python
CT['region'].fillna(CT['region'].mode()[0])
```
