```python
import pandas as pd
```

```python
SM=pd.read_csv('supermarket_messy.csv')
```

```python
SM
```

```python
SM.info()
```

```python
SM['Invoice ID']=SM['Invoice ID'].str.replace('[^0-9]','',regex=True)
```

```python
SM['State']=SM['State'].str.replace('Ohio','OH', regex=False)
```

```python
SM['State']=SM['State'].str.replace('Texas','TX', regex=False)
```

```python
SM['State']=SM['State'].str.upper()
```

```python
SM['City']=SM['City'].str.title()
```

```python
SM['Customer type']=SM['Customer type'].str.lower()
```

```python
SM['gross income']=SM['gross income'].str.replace('gi','',regex=False)
```

```python
SM['Unit price']=SM['Unit price'].str.replace('$','',regex=False)
```
