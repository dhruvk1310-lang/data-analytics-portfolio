```python
ABNB= pd.read_csv('airbnb_ny2019.csv')
```

```python
ABNB
```

```python
ABNB['last_review']=pd.to_datetime(ABNB['last_review'], dayfirst=False, errors='coerce')
```

```python
ABNB.info()
```

```python
ABNB1=ABNB[(ABNB['neighbourhood_group']!='Brooklyn')&(ABNB['last_review']>'2019-06-30')]
```

```python
ABNB1
```

```python
ABNB1['room_type'].value_counts()
```

```python
ABNB1.to_csv('airbnb_subset1.csv')
```

```python
ABNB2=ABNB[(ABNB['neighbourhood_group']=='Manhattan')&(ABNB['number_of_reviews']>400)]
```

```python
ABNB2
```

```python
ABNB2['price'].mean()
```

```python
ABNB2.to_csv('airbnb_subset2.csv')
```
